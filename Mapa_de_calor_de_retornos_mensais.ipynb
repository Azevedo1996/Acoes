```python
import quantstats
import yfinance
import pandas
```

```python
acao = "ITUB3.SA"
cotacoes = yfinance.download(acao, interval='1mo')['Adj Close']
retornos = cotacoes.pct_change().dropna()
quantstats.extend_pandas()
retornos.plot_monthly_heatmap()
```
